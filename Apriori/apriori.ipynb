```python
import numpy as np
import pandas as pd
```

```python
CAMINHO = "C:\\Users\\Nitro\\Desktop\\Redes-Neurais\\compras.csv"
```

```python
df = pd.read_csv(CAMINHO, header=None)
df.replace(np.nan, 0, inplace=True) # inplace = True substitui no próprio dataframe
df.head(10)
```

```python
def removeAllZerosFromList(l):
    return list(filter(lambda x: x != 0, l))
```

```python
list_all_transactions = []

for index, row in df.iterrows():
    listOfTransaction = row.values.tolist()
    listOfTransaction = removeAllZerosFromList(listOfTransaction)

    list_all_transactions.append(listOfTransaction)

print(len(list_all_transactions))
print(list_all_transactions[0])
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(list_all_transactions).transform(list_all_transactions)
df = pd.DataFrame(te_ary, columns=te.columns_) # Adiciona o título da coluna

df.head(10)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True) # Ocorrência dos produtos

frequent_itemsets.sort_values(by=["support"], ascending=False)

```

```python
from mlxtend.frequent_patterns import association_rules
# Lift é a melhor métrica
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold = 0.5)

rules.sort_values(by=["lift"], ascending=False).drop(["antecedent support", "consequent support", "leverage", "conviction"], axis = 1)
```
