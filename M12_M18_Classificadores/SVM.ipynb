```python
from logisticregression2 import LogisticRegression
from knn2 import KNN
from svm import SVM

import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
caminho = "C:\\Users\\Nitro\\Desktop\\Redes-Neurais\\titanic.csv"
```

```python
def getAccuracy(confusionMatrix):
    accuracy = (confusionMatrix[0][0] + confusionMatrix[1][1]) / (confusionMatrix[0][0] + confusionMatrix[0][1] +confusionMatrix[1][0]+confusionMatrix[1][1])
    return accuracy * 100
```

```python
rlArray = []
for i in range(0, 20):
    cmLR = LogisticRegression.computeExample(filename = caminho)
    rlArray.append(getAccuracy(cmLR))
print("Média da Regressão Logística: %.2f" % np.mean(rlArray))
print("Desvio Padrão da Regressão Logística: %.2f" % np.std(rlArray))

knnArray = []
for i in range(0, 20):
    cmKnn = KNN.computeExample(caminho)
    knnArray.append(getAccuracy(cmKnn))
print("\nMédia do KNN: %.2f" % np.mean(knnArray))
print("Desvio Padrão do KNN: %.2f" % np.std(knnArray))

svmArray = []
for i in range(0, 20):
    cmSVM = SVM.computeExample(caminho, "linear", 0)
    svmArray.append(getAccuracy(cmSVM))
print("\nMédia do SVM Linear: %.2f" % np.mean(svmArray))
print("Desvio Padrão do SVM Linear: %.2f" % np.std(svmArray))

svmPoly3Array = []
for i in range(0, 20):
    cmSVMP3 = SVM.computeExample(caminho, "poly", 3)
    svmPoly3Array.append(getAccuracy(cmSVMP3))
print("\nMédia do SCM Poly 3: %.2f" % np.mean(svmPoly3Array))
print("Desvio Padrão do SCM Poly 3: %.2f" % np.std(svmPoly3Array))

svmPoly4Array = []
for i in range(0, 20):
    cmSVMP4 = SVM.computeExample(caminho, "poly", 4)
    svmPoly4Array.append(getAccuracy(cmSVMP4))
print("\nMédia do SVM Poly 4: %.2f" % np.mean(svmPoly4Array))
print("Desvio Padrão do SVM Poly 4: %.2f" % np.std(svmPoly4Array))

svmGaussArray = []
for i in range(0, 20):
    cmSVMG = SVM.computeExample(caminho, "rbf", 0)
    svmGaussArray.append(getAccuracy(cmSVMG))
print("\nMédia do SVM Gaussiano: %.2f" % np.mean(svmGaussArray))
print("Desvio Padrão do SVM Gaussiano: %.2f" % np.std(svmGaussArray))

import matplotlib.pyplot as plt
plt.plot(svmPoly3Array, "r-", knnArray, "g--", svmGaussArray, "b^")
plt.ylabel("Acurácia")
plt.xlabel("Tentativas")
plt.show()
```
