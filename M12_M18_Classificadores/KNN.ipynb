```python
import preprocessing as pre
import logisticregression as lr
import knn as k

import numpy as np
import pandas as pd
```

```python
caminho = "C:\\Users\\Nitro\\Desktop\\Redes-Neurais\\titanic.csv"
```

```python
def getAccuracy(confusionMatrix):
    accuracy = (confusionMatrix[0][0] + confusionMatrix[1][1]) / (confusionMatrix[0][0] + confusionMatrix[0][1] + confusionMatrix[1][0] + confusionMatrix[1][1])
    return accuracy * 100
```

```python
rlArray = []
for i in range(0, 50):
    confusionMatrixR = lr.computeLogisticRegressionExample(caminho)
    rlArray.append(getAccuracy(confusionMatrixR))

knnArray = []
for i in range(0, 50):
    confusionMatrixKNN = k.computeKNNExample(caminho)
    knnArray.append(getAccuracy(confusionMatrixKNN))

import matplotlib.pyplot as plt
plt.plot(rlArray, "r-", knnArray, "g-")
plt.ylabel("Acurácia")
plt.xlabel("Iterações")
plt.show()

print("Média e desvio padrão da regressão Logística: ")
print("%.2f" % np.mean(rlArray), "%.2f" % np.std(rlArray))

print("Média e desvio padrão do KNN: ")
print("%.2f" % np.mean(knnArray), "%.2f" % np.std(knnArray))
```
