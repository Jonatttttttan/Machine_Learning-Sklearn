```python
from logisticregression2 import LogisticRegression
from knn2 import KNN
from svm import SVM
from naivebayes import NaiveBayes
from decisiontree import DecisionTree
from randomforest import RandomForest

import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
caminho = "C:\\Users\\Nitro\\Desktop\\Redes-Neurais\\titanic.csv"
```

```python
def getAccuracy(confusionMatrix):
    accuracy = (confusionMatrix[0][0] + confusionMatrix[1][1]) / (confusionMatrix[0][0] + confusionMatrix[0][1] + confusionMatrix[1][0] + confusionMatrix[1][1])
    return accuracy * 100
```

```python
rlArray = []
for i in tqdm(range(0, 20)):
    cmLR = LogisticRegression.computeExample(caminho)
    rlArray.append(getAccuracy(cmLR))
print("Média da regressão Logística: %.2f" % np.mean(rlArray))
print("Desvio Padrão da regressão logística: %.2f" % np.std(rlArray))

knnArray = []
for i in tqdm(range(0, 20)):
    cmKnn = KNN.computeExample(caminho)
    knnArray.append(getAccuracy(cmKnn))
print("\nMédia do KNN: %.2f" % np.mean(knnArray))
print("Desvio padrão do KNN: %.2f" % np.std(knnArray))

svmGaussArray = []
for i in tqdm(range(0, 20)):
    cmSVMG = SVM.computeExample(caminho, "rbf", 0)
    svmGaussArray.append(getAccuracy(cmSVMG))
print("\nMédia do SVM Gaussiano: %.2f" % np.mean(svmGaussArray))
print("Desvio padrão do SVM Gaussiano: %.2f" % np.std(svmGaussArray))

svmNBArray = []
for i in tqdm(range(0, 20)):
    cmNB = NaiveBayes.computeExample(caminho)
    svmNBArray.append(getAccuracy(cmNB))
print("\nMédia do NB: %.2f" % np.mean(svmNBArray))
print("Desvio padrão do NB: %.2f" % np.std(svmNBArray))

DTArray = []
for i in tqdm(range(0, 20)):
    cmDT = DecisionTree.computeExample(caminho)
    DTArray.append(getAccuracy(cmDT))
print("\nMédia da DT: %.2f" % np.mean(DTArray))
print("Desvio Padrão da DT: %.2f" % np.std(DTArray))

RFArray = []
for i in tqdm(range(0, 20)):
    cmRF = RandomForest.computeExample(caminho)
    RFArray.append(getAccuracy(cmRF))
print("\nMédia da RF: %.2f" % np.mean(RFArray))
print("Desvio padrão da RF: %.2f" % np.std(RFArray))

import matplotlib.pyplot as plt
plt.plot(rlArray, "r-", RFArray, "g--", svmGaussArray, "b^")
plt.ylabel("Acurácia")
plt.xlabel("Tentativas")
plt.show()
```
