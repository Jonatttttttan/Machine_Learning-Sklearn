```python
def runPolynomialRegressionExample(filename):
    start_time = time.time()
    X, y, csv = pre.loadDataset(filename)
    elapsed_time = time.time() - start_time
    print("Load Dataset:n%.2f" % elapsed_time, "segundos")

    start_time = time.time()
    print("Aqui")
    linearRegressor = rl.computeLinearRegressionModel(X, y)
    elapsed_time = time.time() - start_time
    print("Aqui2")
    elapsed_time = time.time() - start_time
    print("Linear Regression Model:n%.2f" % elapsed_time, "segundos")

    start_time = time.time()
    XPoly, polyLinearRegressor = plr.computePolynomialLinearRegressionModel(X, y, 15)
    elapsed_time = time.time() - start_time
    print("Polynomial Regression Model:n%.2f" % elapsed_time, "segundos")

    rl.showPlot(X, y, linearRegressor)
    plr.showPlot(X, y, X, polyLinearRegressor.predict(XPoly))
```

```python

runPolynomialRegressionExample(caminho)
```
