```python
def runRandomForestRegressionExample(filename):
    start_time = time.time()
    X, y, csv = pre.loadDataset(filename)
    elapsed_time = time.time() - start_time
    print("Load Dataset: %.2f" % elapsed_time, "segundos")

    start_time = time.time()
    rfModel = rf.computeRandomForestRegressionModel(X, y, 300)
    elapsed_time = time.time() - start_time
    print("Compute Random Forest Regression: %.2f" % elapsed_time, "segundos")

    rf.showPlot(X, y, X, rfModel.predict(X))

    XGrid = np.arange(min(X), max(X), 0.01)
    XGrid = XGrid.reshape((len(XGrid), 1))
    rf.showPlot(X, y, XGrid, rfModel.predict(XGrid))
```

```python
runRandomForestRegressionExample(caminho)
```
